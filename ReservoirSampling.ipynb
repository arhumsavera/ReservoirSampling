

# Implementing Reservoir Sampling with reservoir size 1

```python
#Arhum Savera
import random
import matplotlib.pyplot as plt

def sampleReservoir(n):
    selected= [0] * 100
    for j in range(n):
        for i in range(0,100): #stream items one by one
            if random.random() < 1.0 / (i+1):
                reservoir=i; #discard the previous element and store the current 
                #print("{} : {}".format("selected ", i))
        selected[reservoir]+=1
    return reservoir,selected

def plotResults(selected,n):
    plt.bar(list(range(0,100)), selected, align='center', color='b', alpha=.8) 
    plt.title("{} iterations of Reservoir Sampling".format(n))
    plt.xlabel('Item Number')
    plt.ylabel('Number of times selected')
    plt.show()
    return
```

# Reporting	the	item sampled in	one	run	of	the	algorithm

```python
res, sel= sampleReservoir(1)
print("{} {}".format(res, "was selected"))
```

# Repeating the algorithm 1000, 10000 and 100000 times and plotting results

```python
for n in (1000,10000,100000):
    res, sel= sampleReservoir(n)
    plotResults(sel,n)
```
